# Visualizing Data

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# Creates a line graph, years on 'x' axis, gdp on 'y' axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# adds a title
plt.title("Nominal GDP")

# adds a label on 'y' axis
plt.ylabel("$ in billions")

# shows graph
plt.show()
```

## Bar graph

```python
# A bar graph is a good choice when you want to show how some quantities vary between a particular set of itens
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Ghandi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# bars with default size of 0.8, so we add 0.1 to the coordinates on the left so each bar will be centralized
xs = [i + 0.1 for i, _ in enumerate(movies)]

# the graph bars with 'x' coordinates on the left [xs], heights [num_oscars]
plt.bar(xs, num_oscars)

plt.ylabel("# Awards")
plt.title("My favorite movies")

# names the 'x' axis with movies titles on the central bar
plt.xticks([i + 0.1 for i, _ in enumerate(movies)], movies)

plt.show()
```

## Histogram 

```python
# A bar graph can be a good choice to create histograms of loaded numerical values, to explore visually how 
# the values are distributted
```

```python
from collections import Counter

# A bar graph can be a good choice to create histogram graphs of loaded numerical values, to visually explore 
# how the values are distributed
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10*10 # Floor Division (also called Integer Division)
histogram = Counter(decile(grade) for grade in grades)

plt.bar([x for x in histogram.keys()],
        histogram.values(),              # gives each bar the right height
        8)                               # gives each bar a wifth of 8

plt.axis([-5, 105, 0, 5])                # 'x' axis of -5 up to 105,
                                         # 'y' axis of 0 up to 5
    
plt.xticks([10*i for i in range(11)])    # 'x' axis labels in 0, 10, ..., 100
plt.xlabel("Decil")
plt.ylabel("# Students")
plt.title("Test 1 distribution of grades")
plt.show()
```

## Line Graph

```python
# Line graphs are a good choice to show tendencies
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# We can make multiple calls for plt.plot to show multiple series of the same graph
plt.plot(xs, variance,     'g-', label='variance') # solid green line
plt.plot(xs, bias_squared, 'r-', label='bias^2') # red pointline
plt.plot(xs, variance,     'b:', label='total error') # blue dotted line

# Due to we assigned labels for each series, we have a free label
# loc=9 means 'top enter'
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.title("Commitment between Polarization and Variance")
plt.show()
```

## Scatter Plots

```python
# A scatter plot is the right choice to visualize the relationship between two pairs of data set. 
# E.g.: the relationship between the number of friends your user has and the number of minutes 
# they spend on the website per day

friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

# give a name for each position
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, 
                xy = (friend_count, minute_count),   # put the label in the respective position
                xytext = (5, -5),                    # but compensates a little
                textcoords='offset points')
    
plt.title("Daily minutes vs. Number of Friends")
plt.xlabel("# of Friends")
plt.ylabel("Daily minutes spent on the website")
plt.show()
```
